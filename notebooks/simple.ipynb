```python
import numpy as np
from rebin1d import (AccumulatedInterpNN,
                     AccumulatedInterpDrizzle,
                     AccumulatedInterpLinear)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = [0, 1, 2, 3, 4]
y = [1, 3, 2, 4, 1.5]

dx = 0.05
xx = np.arange(-1., 5., dx)
xc = 0.5 * (xx[:-1] + xx[1:])

intp = AccumulatedInterpNN(x, y)

yy_nn = intp.rebin(xx) / dx
```

```python
fig, ax = plt.subplots(clear=True)
ax.plot(x, y, "o")
ax.plot(xc, yy_nn, ".-")
```

```python
intp = AccumulatedInterpDrizzle(x, y, drizzle=0.6)

yy_drizzle = intp.rebin(xx) / dx

fig, ax = plt.subplots(clear=True)
ax.plot(x, y, "o")
ax.plot(xc, yy_nn, ".-")
ax.plot(xc, yy_drizzle, ".-")
```

```python
# for Liner interpolation, no extrapolation is allowed for now.
dx = 0.1
xx = np.arange(0.01, 4., dx)
xc = 0.5 * (xx[:-1] + xx[1:])

intp = AccumulatedInterpLinear(x, y)
yy_linear = intp.rebin(xx) / dx
```

```python
fig, ax = plt.subplots(clear=True)
ax.plot(x, y, "o")
ax.plot(xc, yy_linear, ".-")
```
